## MNIST - DE

```python
# load MNIST data without tensorflow
from sklearn.datasets import fetch_openml

# Load MNIST from OpenML
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist.data.values, mnist.target.values
X = X / 255.0  # Normalize pixel values to [0, 1]
y = y.astype(int)
y = jax.nn.one_hot(y, 10)

# Split into training and testing datasets
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]
```

```python
# Fit the model
model = DeepEnsemble(input_dim=X_train.shape[1], hidden_units=200, output_dim=10, num_models=20, model_type='mnist_mlp')
path = '/Users/pgarc/projects/AdvReg/src/models/weights/DE/mnist'
try:
    model.load(path)
except FileNotFoundError:
    print('Not model found, fitting the model') 
    model.fit(
        X_train,
        y_train,
        300,
        batch_size=2048
    )
    model.save(path)
```

```python
# Load notMNIST data without labels
import os
from PIL import Image
import numpy as np

# Ruta de la carpeta con imágenes
folder_path = '../data/notMNIST_small'

# Lista para almacenar las imágenes
images = []

for dirpath, dirnames, filenames in os.walk(folder_path):
    for filename in filenames:
        if filename.endswith('.png'):
            img_path = os.path.join(dirpath, filename)
            img = Image.open(img_path)
            img_gray = img.convert('L')
            img_array = np.array(img_gray)
            images.append(img_array)

X_notmnist = np.array(images).reshape(-1, 28*28) / 255.0
```

## Trying to break them

```python
# Attack notMNIST data to lower the entropy of the predictive distribution
path = '/Users/pgarc/projects/AdvReg/src/models/weights/DE/mnist'
model = DeepEnsemble(input_dim=X_train.shape[1], hidden_units=200, output_dim=10, num_models=20, model_type='mnist_mlp')
model.load(path)
```

```python
def entropy(x, pred):
    pred = pred.mean(axis=0)
    pred += 1e-8
    entr = - (pred * jnp.log(pred)).sum() / jnp.log(2)
    return entr
```

```python
# Y is a 10-dimensional vector with probabilities. Highest entropy is when all probabilities are equal
obj_appd = numpyro.distributions.Dirichlet(jnp.ones(10))

unattacked_entropies = []
for x in tqdm(X_test[:80]):
    x = jax.numpy.array(x.reshape(1, -1))
    preds = model.sample_predictive_distribution(jax.random.PRNGKey(0), x, 20)
    unattacked_entropies.append(entropy(x, preds))
sns.kdeplot(jnp.array(unattacked_entropies), clip=(0, 10))

def process(x, eps):
    x = jax.numpy.array(x.reshape(1, -1))
    x_adv, _ = wasserstein_attack(model, x, appd=obj_appd, rng=jax.random.PRNGKey(0), epsilon=eps, num_samples=20)  
    preds = model.sample_predictive_distribution(jax.random.PRNGKey(0), x_adv, 20)
    return entropy(x_adv, preds)

epsilons = [.2, .5, 1]
for eps in epsilons:
    #entropies = Parallel(n_jobs=8)(delayed(process)(x, eps) for x in tqdm(X_test[:80]))
    entropies = [process(x, eps) for x in tqdm(X_test[:80])]
    sns.kdeplot(jnp.array(entropies), clip=(0, 10))
    

legend = ['Unattacked'] + ['Epsilon = ' + str(eps) for eps in epsilons]
plt.legend(legend)
```

```python
# Example to plot the attacked image 
x = X_test[0]
x = jax.numpy.array(x.reshape(1, -1))
obj_appd = numpyro.distributions.Dirichlet(jnp.ones(10))
x_adv, _ = wasserstein_attack(model, x, appd=obj_appd, rng=jax.random.PRNGKey(0), epsilon=0.5, num_samples=20)
print('Entropy of the original image:', entropy(x, model.sample_predictive_distribution(jax.random.PRNGKey(0), x, 20)))
print('Entropy of the attacked image:', entropy(x_adv, model.sample_predictive_distribution(jax.random.PRNGKey(0), x_adv, 20)))
# Plot the original image
# plot x and x_adv 
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(x.reshape(28, 28), cmap='gray')
ax[0].axis('off')
ax[0].set_title('Original Image')
ax[1].imshow(x_adv.reshape(28, 28), cmap='gray')
ax[1].axis('off')
ax[1].set_title('Adversarial Image')
plt.show()
```

```python
# Attack notMNIST data to lower the entropy of the predictive distribution
unattacked_entropies = []
for x in tqdm(X_notmnist[:100]):
    x = jax.numpy.array(x.reshape(1, -1))
    preds = model.sample_predictive_distribution(jax.random.PRNGKey(0), x, 20)
    unattacked_entropies.append(entropy(x, preds))
sns.kdeplot(jnp.array(unattacked_entropies), clip=(0, 10))

def process(x, eps):
    # Y is a 10-dimensional vector with probabilities. Lowest entropy is when one probability is 1 and the rest are 0
    x = jax.numpy.array(x.reshape(1, -1))
    pred_index = model.sample_predictive_distribution(jax.random.PRNGKey(0), x, 20).mean(axis=0).argmax()
    obj_appd = numpyro.distributions.Dirichlet(jnp.eye(10)[pred_index])
    x_adv, _ = wasserstein_attack(model, x, appd=obj_appd, rng=jax.random.PRNGKey(0), epsilon=eps, num_samples=20)
    preds = model.sample_predictive_distribution(jax.random.PRNGKey(0), x_adv, 20)
    return entropy(x_adv, preds)

for eps in epsilons:
    entropies = Parallel(n_jobs=8)(delayed(process)(x, eps) for x in tqdm(X_notmnist[:80]))
    sns.kdeplot(jnp.array(entropies), clip=(0, 10))


legend = ['Unattacked'] + ['Epsilon = ' + str(eps) for eps in epsilons]
plt.legend(legend)
```
